```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

vx=[]
#21 for 20 epochs
for i in range(0,21):
    vx.append(i)
print(vx)
vy=[0.0,0.53,0.56,0.52,0.59,0.62,0.63,0.67,0.69,0.75,0.76,0.73,0.77,0.79,0.83,0.86,0.87,0.91,0.88,0.93,0.95]

tx=[]
for i in range(0,21):
    tx.append(i)
ty=[0.0,0.43,0.46,0.49,0.54,0.55,0.52,0.57,0.62,0.66,0.69,0.64,0.69,0.73,0.76,0.79,0.81,0.83,0.87,0.92,0.97]


# Plot training accuracy versus validation accuracy
plt.figure(figsize=(10, 6))
plt.plot(vx,vy, label='Validation Accuracy',marker='o', color='b')
plt.plot(tx,ty, label='Training Accuracy',marker='o', color='r')
plt.title('Training Accuracy vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(False)

# Save the plot as SVG
plt.savefig('Accuracy_Plot.svg', format='svg', bbox_inches='tight')
plt.show()
```
